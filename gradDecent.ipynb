```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def y_function(x):
    return np.sin(x)
def y_derivative(x):
    return np.cos(x)
```

```python
x = np.arange(-5,5,0.1)
y = y_function(x)
current_position = (1.5,y_function(1.5))
learning_rate = 0.01
plt.plot(x,y)
plt.scatter(current_position[0],current_position[1],color = "red")
plt.show()
```

```python
for _ in range(1200):
    new_x = current_position[0] - learning_rate * y_derivative(current_position[0])
    new_y = y_function(new_x)
    current_position = (new_x,new_y)
    plt.plot(x,y)
    plt.scatter(current_position[0],current_position[1],color = "red")
    plt.pause(0.001)
    plt.clf()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function and gradient
def z_function(x, y):
    return np.sin(5*x) * np.cos(5*y) / 5

def calculate_gradient(x, y):
    return np.cos(5*x) * np.cos(5*y), -np.sin(5*x) * np.sin(5*y)

# Create a grid of x and y values
x = np.arange(-1, 1, 0.05)
y = np.arange(-1, 1, 0.05)
X, Y = np.meshgrid(x, y)
Z = z_function(X, Y)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

# Initialize the gradient descent parameters
current_pos = (0.7, 0.4)
learning_rate = 0.01

# Gradient descent loop
for _ in range(1000):
    X_d, Y_d = calculate_gradient(current_pos[0], current_pos[1])
    X_new, Y_new = current_pos[0] - learning_rate * X_d, current_pos[1] - learning_rate * Y_d
    Z_new = z_function(X_new, Y_new)
    
    # Update current position
    current_pos = (X_new, Y_new)
    #do for surfcae  wit zorder = 0
    # Clear previous scatter plot and plot new point
    ax.scatter(current_pos[0], current_pos[1], z_function(current_pos[0], current_pos[1]), color='red',zorder =1)

    # Update the plot
    plt.draw()
    plt.pause(0.001)

plt.show()
```
